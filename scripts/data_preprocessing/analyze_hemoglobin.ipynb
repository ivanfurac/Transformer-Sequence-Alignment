```python
from Bio import SeqIO
```

```python
lengths = []
i = 0

for seq in SeqIO.parse('./fasta/hemoglobin.fasta', 'fasta'):
    lengths.append(len(seq.seq))
    i += 1
    
print(i)
```

```python
i = 0
for len in lengths:
    if(len >= 100 and len <= 200):
        i+=1
        
print(i)
```

```python
limits = []
counters = []

for limit in range(100, 6300, 100):
    limits.append(limit)
    counter = 0
    for length in lengths:
        if(length <= limit):
            counter = counter + 1
    counters.append(counter)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))

plt.plot(limits, counters, color='orange')
plt.xlabel('Duljina slijeda (broj aminokiselina u slijedu)')
plt.ylabel('Broj sljedova')

plt.show()
```
